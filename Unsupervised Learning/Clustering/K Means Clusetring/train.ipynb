```python
import numpy as np
from sklearn.datasets import make_blobs

from k_means import KMeans
```

### Constants and hyperparameters

```python
centers = 3 # number of clusters
n_samples = 500 # number of samples to generate
n_features = 2 # number of features for each sample
seed = 42 # seed for reproducibility
num_iterations = 150 # number of iterations for the KMeans algorithm
```

### Data loading

```python
# Creating the dataset
X, y = make_blobs(centers=centers, n_samples=n_samples, n_features=n_features, shuffle=True, random_state=seed) # type: ignore

# Extracting the number of clusters
clusters = len(np.unique(y))

# Displaying the number of clusters
print("Number of clusters:", clusters)
```

### Building the model

```python
# Initializing the KMeans class
model = KMeans(k=clusters, max_iterations=num_iterations, plot_steps=False)
```

### Training the model

```python
# Extracting the clusters
y_pred = model.predict(X)

# Plotting the clusters
model.plot()
```
