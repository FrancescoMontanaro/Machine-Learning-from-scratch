```python
import numpy as np
from sklearn import datasets
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split

from src.linear_regression import LinearRegression
```

### Constants and hyperparameters

```python
train_test_split_pct = 0.2 # percentage of data to use for testing
n_features = 1 # number of features
n_samples = 1000 # number of samples to generate
epochs = 1000 # number of epochs to train the model
noise = 20 # noise level to add to the data
seed = 1234 # seed for random number generation
learning_rate = 0.01 # learning rate for the model0
```

### Data loading

```python
# Loading the dataset
X, y = datasets.make_regression(n_samples=n_samples, n_features=n_features, noise=noise, random_state=seed) # type: ignore

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=train_test_split_pct, random_state=seed)
```

### Building the model

```python
# Creating the linear regression model
model = LinearRegression(alpha=learning_rate, epochs=epochs)
```

### Training the model

```python
# Fitting the model
model.fit(X_train, y_train)
```

### Evaluating the model

```python
# Predicting the labels of the test set
y_pred = model.predict(X_test)
```

```python
# Computing the Mean Squared Error (MSE) of the predictions
mse = np.mean((y_pred - y_test) ** 2)

# Printing the MSE
print("Mean Squared Error:", mse)
```

```python
# Plotting the predictions
plt.scatter(X_test, y_test, s=5)
plt.plot(X_test, y_pred, color="red")
plt.show()
```
